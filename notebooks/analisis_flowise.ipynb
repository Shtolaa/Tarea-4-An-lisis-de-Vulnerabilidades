Notebook base para analizar el dataset generado por el pipeline del repositorio.

```python
from pathlib import Path
import pandas as pd

ROOT = Path('..')
findings = pd.read_csv(ROOT / 'data' / 'findings.csv')
repos = pd.read_csv(ROOT / 'data' / 'top_repos.csv')
findings.head()
```

## Dimension cuantitativa

```python
findings.groupby(['repo', 'dimension']).size().unstack(fill_value=0)
```

```python
findings.groupby(['scanner', 'severity']).size().sort_values(ascending=False)
```

```python
dependency_findings = findings[findings['dimension'] == 'dependencies']
dependency_findings.groupby(['package_name', 'severity']).size().sort_values(ascending=False).head(20)
```
